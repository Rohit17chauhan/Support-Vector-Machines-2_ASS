Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

```python
"""In Support Vector Regression (SVR), the parameter epsilon (ε) defines the width of the epsilon-insensitive tube around the regression line (or hyperplane). This tube specifies the margin of tolerance where no penalty is given for errors in predictions. Increasing the value of epsilon has a direct impact on the number of support vectors in the SVR model.

Effects of Increasing Epsilon on Support Vectors:
Wider Margin:

As epsilon increases, the epsilon-insensitive tube widens. This means that more data points fall within the margin without incurring a penalty for prediction errors.
Fewer Support Vectors:

When the tube is wider, fewer data points are likely to be outside this margin. Consequently, fewer points will be classified as support vectors. Support vectors are the data points that lie outside the epsilon-insensitive tube and thus are critical in defining the regression function.
As a result, increasing epsilon typically leads to a decrease in the number of support vectors because more points fall within the tube, reducing the points that influence the model.
Bias-Variance Trade-off:

A larger epsilon can lead to a simpler model with higher bias but lower variance, as the model becomes less sensitive to the noise in the data. This can help in avoiding overfitting.
Conversely, a smaller epsilon means a tighter fit to the data, potentially capturing more variations and leading to more support vectors, but it may also increase the risk of overfitting the model to the training data.
Example:
Low Epsilon: If epsilon is set to a low value, only a few points may fall within the epsilon-insensitive tube, leading to many support vectors being used to define the regression line. The model is more sensitive to each data point.

High Epsilon: If epsilon is increased, many data points will likely fall within the tube, resulting in fewer support vectors. The regression line may not fit the training data as closely, which can help generalize better to unseen data.

"""
```

Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter
affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works
and provide examples of when you might want to increase or decrease its value?

```python
"""In Support Vector Regression (SVR), the performance of the model is significantly influenced by the choice of kernel function and the values of several key parameters: C, epsilon (ε), and gamma (γ). Here’s an overview of how each parameter works and how adjustments can impact performance:

1. Kernel Function:
The kernel function defines the way data is transformed into a higher-dimensional space to enable non-linear regression.

Types of Kernel Functions:

Linear Kernel: Suitable for linearly separable data. It has a simple structure but may underfit complex data.
Polynomial Kernel: Can capture interactions between features, making it useful for polynomial relationships.
Radial Basis Function (RBF) Kernel: Very flexible and widely used for non-linear problems. It can adapt well to complex data structures.
Effect on Performance:

Choosing the right kernel function is crucial. For example, if your data is inherently linear, a linear kernel might suffice. However, for non-linear relationships, an RBF kernel may provide better performance.
When to Adjust:

If the model underfits, you might switch to a more complex kernel (e.g., RBF) to capture non-linear relationships.
If the model overfits, switching to a simpler kernel (e.g., linear) might be beneficial.
2. C Parameter:
The C parameter controls the trade-off between achieving a low training error and a low testing error (generalization).

Functionality:

A small value of C allows some errors in predictions, which can lead to a wider margin and potentially fewer support vectors. This can help prevent overfitting but may lead to underfitting.
A large value of C penalizes errors heavily, leading to a tighter fit to the training data, which can reduce bias but increase variance (risk of overfitting).
Effect on Performance:

A larger C results in a model that closely follows the training data, potentially capturing noise.
A smaller C results in a more generalized model that may not capture all the data intricacies.
When to Adjust:

If the model overfits, consider decreasing C to allow more flexibility.
If the model underfits, increasing C can help tighten the fit.
3. Epsilon Parameter (ε):
The epsilon (ε) parameter defines the width of the epsilon-insensitive tube around the predicted values.

Functionality:

A smaller epsilon means the model pays closer attention to small errors, leading to more support vectors and a tighter fit to the data.
A larger epsilon creates a wider tube, allowing more data points to be within the margin without being penalized, which may result in fewer support vectors.
Effect on Performance:

A smaller epsilon may increase model complexity and sensitivity to noise, which can lead to overfitting.
A larger epsilon may simplify the model, reducing sensitivity to noise and potentially improving generalization.
When to Adjust:

If the model is too sensitive to noise, increasing epsilon can help create a more robust model.
If you need a precise fit, decreasing epsilon allows the model to consider more data points as significant.
4. Gamma Parameter (γ):
The gamma (γ) parameter defines how much influence a single training example has on the decision boundary.

Functionality:

In the context of RBF and polynomial kernels, a high value of gamma means that each training point has a significant influence on the decision boundary, leading to more complex models that can closely fit the training data.
A low value of gamma results in a smoother decision boundary that is less sensitive to individual points, which can help generalization.
Effect on Performance:

A high gamma can lead to overfitting as the model may adapt too closely to the training data.
A low gamma can result in underfitting, as the model may be too simplistic to capture the underlying patterns.
When to Adjust:

If the model overfits, consider decreasing gamma to smooth out the decision boundary.
If the model underfits, increasing gamma can help it capture more complex patterns."""
```

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
